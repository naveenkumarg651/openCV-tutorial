# Image writing


```python
import numpy as np
import cv2
```

```python
image=np.random.rand(256,256) # generating 256x256 image using random function
```

```python
#displaying random image
cv2.imshow('random_image',image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#writing image with name "random image" in .jpg format
cv2.imwrite("random_image.jpg",image)
```

# Greyscale conversion

```python
image=cv2.imread("numbers.jpg") #reading image
cv2.imshow("normal image",image) # displaying normal image
cv2.waitKey()
cv2.destroyAllWindows()
grey_scale=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)# converting to grey scale image
cv2.imshow("greyscale_image",grey_scale)#displaying grey scale image
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#fastest way of doing this 
gimage=cv2.imread("numbers.jpg",0)
cv2.imshow("greyscale image",gimage)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Color Spaces

```python
image=cv2.imread("deepcopy.jpg")
cv2.imshow("deepcopy",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
B,G,R=cv2.split(image)
print(image.shape,B.shape,G.shape,R.shape)
```

```python
zeros=np.zeros((350,350),dtype="uint8")

cv2.imshow("R",cv2.merge([zeros,zeros,R]))
cv2.imshow("B",cv2.merge([B,zeros,zeros]))
cv2.imshow("G",cv2.merge([zeros,G,zeros]))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Plotting histogram of  image channels

```python
import matplotlib.pyplot as plt
image=cv2.imread("deepcopy.jpg")
histogram=cv2.calcHist([image],[1],None,[256],[0,256])#image,channel,mask,histlimit,hist range
plt.hist(histogram)
plt.show()
```

# BGR 2 HSV conversion

```python
hsv_image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
plt.imshow(hsv_image)
plt.show()
```
